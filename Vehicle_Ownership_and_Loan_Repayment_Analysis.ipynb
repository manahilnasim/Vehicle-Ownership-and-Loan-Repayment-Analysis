```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
# Load the application_train.csv file
df = pd.read_csv('application_train.csv')
```

CONTRACT STATUS

---


Contract status is a good starting variable as it gives accesss to information regarding the aplicants loan history. Contract status cannot be the main source for loan consideration but it is a good starting point to dig deeper into other factors such as income, employment and family history.

```python
prev_app_df = pd.read_csv('previous_application.csv')
status_counts = prev_app_df['NAME_CONTRACT_STATUS'].value_counts()
status_counts.plot(kind='bar')
plt.title('Distribution of NAME_CONTRACT_STATUS')
plt.xlabel('NAME_CONTRACT_STATUS')
plt.ylabel('Frequency')
plt.show()
```

OWNING A VEHICLE

---

This variable gives insights relating to the applicants assets and loan repayment ability. The applicants who own vehicles can be seen as ones that may be able to pay off their loans considering they have assets.

```python
sns.catplot(x="FLAG_OWN_CAR", col="TARGET", kind="count", data=df, hue="FLAG_OWN_CAR")
```

AGE OF CLIENT

---

Age is an important factor to take into consideration as if an applicant is nearing the age of retirement, there is a higher possibility of default occuring. On the contrary, if the applicant is very young, there may be risks asscociated with ability of loan repayment.

```python
# Plot the distribution of ages in years
plt.hist(df['DAYS_BIRTH']/-365, edgecolor = 'k', bins = 25)
plt.title('Age of Client'); plt.xlabel('Age (years)'); plt.ylabel('Count')
```

OCCUPATION TYPE

---
Occupation is another variable that can be taken into consideration, as it provides information regarding applicants ability to repay the loan based on income earned. Someone with a higher paying stable occupation can be approved for a loan easier considering there is lower risks associated. However, occupation should not be used in a discriminatory manner as an applicants ability to repay a loan should not be based solely on their occupation.


```python
fig, ax = plt.subplots(figsize=(10,5))
sns.countplot(x='OCCUPATION_TYPE',data=df)
plt.xlabel("Occupation Type")
plt.xticks(rotation=70)
```

OCCUPATION TYPE VS TARGET

---
Based on the plot below it is evident that laborers may have higher need for loans considering the lower income earned. On the other hand, IT and high skill occupations may have higher incomes and hence have less of a need for loans.
Laborers are also more inclined to making payments as a means of keeping up their credit score.


```python
fig, ax = plt.subplots(figsize=(10,5))
sns.countplot(x='OCCUPATION_TYPE',hue='TARGET',data=df)
plt.xticks(rotation=70)
plt.xlabel("Occupation Type")
```

GENDER VS TARGET

---

Gender is a difficult variable as it should not be used once again in a discriminatory manner for loan approval. However, by viewing trends and default risks it can be helpful for input in a predictive model.

```python
fig,ax = plt.subplots(figsize = (10,5))
sns.countplot(x='CODE_GENDER',data=df,hue='TARGET')
plt.xticks(rotation=70)
plt.xlabel("Gender")
```

REAL ESTATE VS TARGET

---
Owning real estate can be relevant in the field of predictive modeling, as it can be an indicator of financial stability. It can also be used as collateral and a means of indicating equity. Below, it can be seen that there are far more loan applicants who own real estate than those who do not, as this sense of stability provides more opportunity for investments to be made.


```python
fig, ax = plt.subplots(figsize=(8,4))
sns.countplot(x='FLAG_OWN_REALTY',hue='TARGET',data=df)
plt.xticks(rotation=70)
plt.xlabel("Owning Real Estate")
```

Level of Education VS TARGET

---
Education can be an indicator of a borrowers earning potential. Higher levels of education are associated with higher levels of income. However, there are risks associated with higher level of education, considering student loans that are to be paid off.


```python
fig, ax = plt.subplots(figsize=(8,4))
sns.countplot(x='NAME_EDUCATION_TYPE',hue='TARGET',data=df)
plt.xticks(rotation=70)
plt.xlabel("Level of Education")
```

FAMILY STATUS VS TARGET

---
Family status can also be an important factor for predictive modeling. Below we can see that the 'Married' family status tend to have far more needs than other family statuses hence the highest need for a loan. This can range for home improvements to child needs and other emergencies that may arise. This can be an important factor for consideration of applicants loan along with other determining factors.

```python
fig, ax = plt.subplots(figsize=(8,4))
sns.countplot(x='NAME_FAMILY_STATUS',hue='TARGET',data=df)
plt.xticks(rotation=70)
plt.xlabel("Family Status")
```

NUMBER OF CHILDREN VS TARGET

---
This variable can be used to determine the applicants financial burden. The number of children one has can be an indication of the applicants financial obligations and their ability to repay a loan.


```python
fig, ax = plt.subplots(figsize=(8,4))
sns.countplot(x='CNT_CHILDREN',hue='TARGET',data=df)
plt.xticks(rotation=70)
plt.xlabel("Number of Children")
```
